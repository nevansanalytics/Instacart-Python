# Alternative way for Conditional Logic using "loc()" - Preferred Method which Runs Faster and No Warnings

```python
# Defining High-range product
# 'if' implied before the ',' ... 'then' implied after the ','

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Defining Mid-range product
# # 'if' implied before the ',' ... 'then' implied after the ','

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Defining Low-range product
# # 'if' implied before the ',' ... 'then' implied after the ','

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Assigning labels based on the day of the week in the "orders_day_of_week" column.
# If the value is 0 or 1, it's labeled as "Busiest days".
# If the value is 4 or 3, it's labeled as "Slowest days".
# For all other values, it's labeled as "Regularly busy".

result = []

for value in df["orders_day_of_week"]:
    if value in [0, 1]:
        result.append("Busiest days")
    elif value in [3, 4]:
        result.append("Slowest days")
    else:
        result.append("Regularly busy")

# Create a new column with the labels
df['busiest_days'] = result

# Check the result
print(df.head())
```

```python
# Check value counts

df['busiest_days'].value_counts(dropna = False)
```

## 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Calculate the order frequency for each hour
order_counts = df['order_hour_of_day'].value_counts()

# Define thresholds for busiest, average, and slowest hours
busiest_threshold = order_counts.max() * 0.7
slowest_threshold = order_counts.max() * 0.3

# Create a new column with labels using loc
df.loc[df['order_hour_of_day'].isin(order_counts[order_counts >= busiest_threshold].index), 'busiest_period_of_day'] = 'Most orders'
df.loc[df['order_hour_of_day'].isin(order_counts[order_counts <= slowest_threshold].index), 'busiest_period_of_day'] = 'Fewest orders'
df['busiest_period_of_day'].fillna('Average orders', inplace=True)

# Print the frequency of each label
print(df['busiest_period_of_day'].value_counts())
```
